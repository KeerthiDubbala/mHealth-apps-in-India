### 1. Released Year

```python
import datetime
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define the colors for each disease
colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# Filter out rows with missing or invalid values in the 'released' column
filtered_df = all_df.dropna(subset=['released'])
filtered_df['released'] = pd.to_datetime(filtered_df['released'], errors='coerce')

# Get unique diseases for grouping
unique_diseases = sorted(filtered_df['disease'].unique())  # Sort diseases alphabetically

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Iterate through unique diseases and plot the distribution of timestamps by year
for i, disease in enumerate(unique_diseases):
    # Filter the dataframe for the current disease
    current_df = filtered_df[filtered_df['disease'] == disease]
    
    # Extract the timestamps for the current disease
    timestamps = current_df['released'].tolist()
    
    # Remove NaT (missing) values
    timestamps = [ts for ts in timestamps if not pd.isnull(ts)]
    
    # Create a dictionary to store the counts for each year
    year_counts = {}
    
    for ts in timestamps:
        year = ts.year
    
        if year in year_counts:
            year_counts[year] += 1
        else:
            year_counts[year] = 1
    
    # Sort the years in ascending order
    sorted_years = sorted(year_counts.items())
    
    # Extract the years and counts for plotting
    years = [year[0] for year in sorted_years]
    counts = [year[1] for year in sorted_years]
    
    # Get the color for the current disease
    color = colors[i % len(colors)]
    
    # Create a line plot for the years with the respective color
    ax.plot(years, counts, marker='o', label=disease, color=color)
    
# Set the x-axis and y-axis labels
ax.set_xlabel("Year")
ax.set_ylabel("Count")

# Set the title and legend
ax.set_title("Distribution of Timestamps by Year for Each Disease")
ax.legend()

# Set the x-axis ticks
ax.set_xticks(years)

# Enable grid lines
ax.grid(True)

# Adjust the layout
plt.tight_layout()

# Save the figure using the save_figure function
save_figure('Distribution of Timestamps by Year for Each Disease')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

# # Load your dataset
# data = pd.read_csv('data.csv')

# Specify the independent variables
X = all_df[['released_year', 'size_mb', 'android_2022', 'price', 'score', 'description_length', 
                               'iap', 'encoded_disease', 'ratings', 'reviews']]

# Specify the dependent variable
y = all_df['installs_2022']

# Add a constant term to the independent variables matrix (required for statsmodels)
X = sm.add_constant(X)

# Perform linear regression
model = sm.OLS(y, X).fit()

# Check assumptions
residuals = model.resid

# 1. Linearity and Independence of Errors
plt.scatter(model.fittedvalues, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Linearity and Independence of Errors')
plt.show()

# 2. Homoscedasticity (Constant Variance)
plt.scatter(model.fittedvalues, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Homoscedasticity (Constant Variance)')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()

# 3. Normality of Residuals
sm.qqplot(residuals, line='s')
plt.title('Normality of Residuals')
plt.show()

# 4. Multicollinearity (VIF)
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif)

# 5. Autocorrelation (Durbin-Watson Test)
from statsmodels.stats.stattools import durbin_watson
print("Durbin-Watson:", durbin_watson(residuals))

# Print regression summary
print(model.summary())
```
